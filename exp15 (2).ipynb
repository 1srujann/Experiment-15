```python
from nltk.tokenize import word_tokenize
text = "What is natural language processing?!"
tokens = word_tokenize(text)
print(tokens)
```

```python
#sentence tokenization
from nltk.tokenize import sent_tokenize
text = "we will learn natural language processing in a short while"
sentences = sent_tokenize(text)
print(sentences)

```

```python
#stop word removal
#stop words such as it,the,and,in are removed.
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
words=word_tokenize(text)
filtered_words =[w for w in words if w.lower() not in stop_words]
print(filtered_words)
```

```python
#stemming
#stemming reduces words to their root form
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
words = ["playing", "helping", "running"]

for w in words:
  print(stemmer.stem(w))
```

```python
#lemmatization
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize("dancing"))
print(lemmatizer.lemmatize("acting"))
print(lemmatizer.lemmatize("worst"))
```
